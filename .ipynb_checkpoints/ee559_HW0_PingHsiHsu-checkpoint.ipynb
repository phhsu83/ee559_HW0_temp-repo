```python
import numpy as np

data_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
data_nparr = np.array(data_list)
print(data_nparr)
```

```python
print("ndim:", data_nparr.ndim)
print("shape:", data_nparr.shape)
print("size:", data_nparr.size)
print("dtype:", data_nparr.dtype)
print("itemsize:", data_nparr.itemsize)
print("data:", data_nparr.data)
```

```python
reshaped = data_nparr.reshape(1, 9)
flattened = data_nparr.flatten()
print("reshaped:", reshaped)
print("flattened:", flattened)
```

```python
print("1D array slice:", data_nparr[0, :2])
print("2D array slice:", data_nparr[:2, 1:])
```

```python
print("argmin:", np.argmin(data_nparr))
print("argmax:", np.argmax(data_nparr))
print("min:", np.min(data_nparr))
print("max:", np.max(data_nparr))
print("mean:", np.mean(data_nparr))
print("sum:", np.sum(data_nparr))
print("std:", np.std(data_nparr))
print("square:", np.square(data_nparr))
print("sqrt:", np.sqrt(np.abs(data_nparr)))
print("abs:", np.abs(data_nparr))
print("exp:", np.exp(data_nparr))
print("sign:", np.sign(data_nparr))
print("mod:", np.mod(data_nparr, 3))
```

```python
arange = np.arange(0, 10, 2)
print("arange:", arange)

ones = np.ones((2, 2))
print("2D ones:", ones)

zeros = np.zeros((2, 2))
print("2D zeros:", zeros)

eye = np.eye(3)
print("3x3 identity matrix:\n", eye)

linspace = np.linspace(0, 10, 5)
print("linspace:", linspace)

concatenate1 = np.array([[1, 2], [3, 4]])
concatenate2 = np.array([[5, 6], [7, 8]])
concat_axis0 = np.concatenate((concatenate1, concatenate2), axis=0)
concat_axis1 = np.concatenate((concatenate1, concatenate2), axis=1)
print("concatenate_axis0", concat_axis0)
print("concatenate_axis1", concat_axis1)
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y)
plt.title("title")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, 400)

y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8, 6))

plt.plot(x, y1, 'r-', label=r'$\sin(x)$')
plt.plot(x, y2, 'b--', label=r'$\cos(x)$')

plt.legend(loc='upper right', fontsize=12)

plt.title("title", fontsize=14)
plt.xlabel("x axis", fontsize=12)
plt.ylabel("y axis", fontsize=12)

plt.grid(True, linestyle='--', alpha=0.6)

plt.text(-8, 0.8, r'$y = \sin(x)$', fontsize=12, color='red')
plt.text(4, 0.5, r'$y = \cos(x)$', fontsize=12, color='blue')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.subplot(2, 1, 1)
plt.plot(x, y1, label="line 1")
plt.legend()

plt.xlabel("x axis", fontsize=12)
plt.ylabel("y axis", fontsize=12)

plt.subplot(2, 1, 2)
plt.plot(x, y2, label="line 2")
plt.legend()

plt.xlabel("x axis", fontsize=12)
plt.ylabel("y axis", fontsize=12)

plt.show()
```

```python
x = np.linspace(-10, 10, 400)

y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8, 6))

plt.plot(x, y1, 'r-', label=r'$\sin(x)$')
plt.plot(x, y2, 'b--', label=r'$\cos(x)$')

plt.title("title", fontsize=14)
plt.xlabel("x axis", fontsize=12)
plt.ylabel("y axis", fontsize=12)

ax = plt.gca()

ax.set_xlim([0, 12])  
ax.set_ylim([-1.5, 1.5])  

ax.set_xlabel("new x axis", fontsize=12)
ax.set_ylabel("new y axis", fontsize=12)

ax.set_title("new title", fontsize=14)

ax.set_xticks([0, 2, 4, 6, 8, 10])
ax.set_yticks([-1, 0, 1])

ax.grid(True, linestyle='--', alpha=0.6)

plt.legend(loc='upper right', fontsize=12)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Salaries.csv')
print(df)
```

```python
sns.scatterplot(data=df, x="salary", y="teamID")
plt.show()
```

```python
sns.lmplot(data=df, x="salary", y="teamID", height=6)
plt.show()
```

```python
sns.catplot(data=df, x="teamID", y="salary", kind="box", height=6)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.relplot(data=df, x="salary", y="teamID", kind="scatter", height=6)
plt.show()
```

```python
sns.boxplot(data=df, x="teamID", y="salary")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.jointplot(data=df, x="salary", y="teamID", kind="scatter")
plt.show()
```
